```python
import os
import pandas as pd
import shutil
```

```python
isic2019_data = pd.read_csv('/home/ubuntu/data/isic-2019/ISIC_2019_Training_GroundTruth.csv')
```

```python
print(isic2019_data.columns)
```

```python
isic2019 = isic2019_data.melt(id_vars = 'image', var_name = 'diagnostic', value_name = 'Flag')
```

```python
isic2019_data.shape
```

```python
isic2019
```

```python
isic2019 = isic2019[isic2019.Flag==1]
```

```python
isic2019 =isic2019.drop(columns=['Flag'])
```

```python
isic2019
```

```python
dest_dir = '/home/ubuntu/data/isic-2019/images_by_class/'
data_dir = '/home/ubuntu/data/isic-2019/ISIC_2019_Training_Input/ISIC_2019_Training_Input/'
```

```python
label = isic2019['diagnostic'].unique().tolist()
print(label)
label_images = []
```

```python
for i in label:
    print(dest_dir + str(i) + '/')
    os.mkdir(dest_dir + str(i))
    sample = isic2019[isic2019['diagnostic'] == i]['image']
    label_images.extend(sample)
    for id in label_images:
        shutil.copyfile((data_dir + '/' + id + '.jpg' ), (dest_dir + str(i) + '/' + id+ '.jpg' ))
    label_images = []
```
